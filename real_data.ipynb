**Guidelines for applying DLS to real neuroimaging data (e.g., HCP data) to jointly estimate whole-brain gradients and parcels (spatial ICA maps).**

```python
import dls 
```

```python
whole_brain = dls.DLSModel(res=0.25) # Sampling rate is set to 0.25
```

```python
DC_train, DC_val = whole_brain.data_to_corr("/home/fs0/bsg097/scratch/Data") # Specify the path to the directory containing only time-series CIFTI files. Half/Half split for train and eval.
```

```python
whole_brain.fit_transform(DC_train, method = 'UMAP', N = 100, g = 2, r = 9, MR = 9) # Joint estimation of gradients and parcels
```

```python
whole_brain.vis_DCs() # Visualising the corresponding connectomes
```

```python
whole_brain.vis_grads(side='left', idx=0) # Visualising gradients
```

```python
whole_brain.vis_parcels(side='left', idx=0) # Visualising parcels
```
